```python
import csv
import pandas as pd
import codecs
import os
import glob
```

```python
# Making a list directory
files = glob.glob('./datasets/water_tweets/gathered_water/*.csv')
#print(files)

un_h2o_tweets_1 = pd.concat([pd.read_csv(f, encoding = 'ISO-8859-1', delimiter = ';') for f in files])
un_h2o_tweets_1 = un_h2o_tweets_1[['category', 'date', 'text', 'retweet_count', 'tweet_id']]
un_h2o_tweets_1 = un_h2o_tweets_1[un_h2o_tweets_1['text'] != 'text']
un_h2o_tweets_1 = un_h2o_tweets_1.drop_duplicates(subset = ['text'], keep = 'first').reset_index()
un_h2o_tweets_1 = un_h2o_tweets_1.iloc[:, 1:]
un_h2o_tweets_1.columns = ['category', 'date', 'tweet', 'retweet_count', 'tweet_id']
print(len(un_h2o_tweets_1))

un_h2o_tweets_1.head()
```
